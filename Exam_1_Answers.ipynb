## Problem 6

* Create a command that shows all of the lines that have a phenotype including the word "root"

```bash
grep "\sroot\s" -n Locus_Germplasm_Phenotype_20130122.txt
```

grep with -n prefixes each line of output with the 1-based line number within its input file to see all of the lines that have a phenotype including the word "root". I do not use -i because I do not want to ignore case distinctions.





## Problem 7

* Create a command that writes the AGI Locus Code for every line that has a phenotype including the word "root" to a file called: **Root-associated-Loci.txt**

```bash
grep "\sroot\s" -n Locus_Germplasm_Phenotype_20130122.txt | grep -iEo "AT[1-5]G[0-9]{5}" > Root-associated-Loci.txt
```

Using grep -o only prints the matched parts of a matching line. It allows to introduce all* the Locus Codes in a new file.


*Although the last loci do not follow the AGI Locus Code, I have only searched for AGI Locus Codes as required by the Problem.








## Problem 8
